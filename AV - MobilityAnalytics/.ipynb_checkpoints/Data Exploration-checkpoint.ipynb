## Missing Value Treatment

```python
missing_data = pd.DataFrame({'Col_Name':list(subm),
                            'Data':0,
                            'Subm':0})
for i in list(subm):
    missing_data.loc[missing_data['Col_Name']==i,'Data'] = sum(data[i].isna())
    missing_data.loc[missing_data['Col_Name']==i,'Subm'] = sum(subm[i].isna())

(missing_data)
```

### Type_of_Cab

```python
print(data.Type_of_Cab.value_counts())
print(subm.Type_of_Cab.value_counts())
sample = data
sample.loc[sample.Type_of_Cab.isna(),'Type_of_Cab'] = 'Others'
```

```python
# sample[['Type_of_Cab','Var3']].groupby('Type_of_Cab').agg(['mean','median', pd.Series.mode, 'std'])
sample[['Type_of_Cab','Surge_Pricing_Type']].groupby('Type_of_Cab').agg(['mean'])
```

##### There is no significant relationship between the type of cab and anyother column, hence replacing the missing values with 'Other'

```python
data = sample
subm.loc[subm.Type_of_Cab.isna(),'Type_of_Cab'] = 'Others'
```
